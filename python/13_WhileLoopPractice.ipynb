### how to check wheather a given is prime or not 

    prime 
    
        a number which divisble by 1 and itself
        
         if a number is not divible by any number other than 1 and itself

    1 -> not prime
    2, 3 -> prime

 2 to num - 1

```python
num = 127
check = 2

while check < num:
    print(f"{num} % {check} = {num%check}")
    if num % check == 0:
        print("Not Prime")
        break
    check += 1
    
# num -> sqrt(num) ?   
```

```python
num = int(input("Enter ay number: "))
flag = True
check = 2
while check <= math.sqrt(num)+1:
    if num % check == 0:
        flag = False
        break
    check += 1
    
if flag:
    print(f"Given number {num} is Prime")
else:
    print(f"Given number {num} is not Prime")
```

```python
start = int(input("start : ")) # 1
end = int(input("end : ")) # 10
num = start
range_ = (start, end)
count = 0
while num <= end:
    flag = True
    if num <= 1:
        flag = False
    elif num <= 3:
        flag = True
    elif num % 2 == 0:
        flag = False
    else:
        check = 3
        while check <= math.sqrt(num)+1:
            if num % check == 0:
                flag = False
                break
            check += 2
    if flag:
        count += 1
        print(num, end=", ")
    num += 1
print(f"\nTotal Prime numbers in range {range_} are {count}")
```

```python
records = [
    # name, maths, chem, phys
    ["sachin", 60, 70, 80],
    ["rajat", 89, 78, 56],
    ["simran", 80, 70, 60],
    ["rahul", 60, 70, 60]
]
```

```python
name = input("student name: ").strip().lower()
i = 0
while i < len(records):
    st = records[i]
    if st[0] == name:
        per = sum(st[1:])/3
        print(f"{name} has scored {per:.2f}% marks.")
        break
    i += 1
else:
    print("Student Record Not Found!")
    
```
