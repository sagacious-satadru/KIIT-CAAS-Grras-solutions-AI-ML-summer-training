```python
def func(value):
    if value!=0:
        return value+func(value-1)
    return 0
```

```python
def square_sum(listNums: list)->float:
    """
        will return square sum of all numbers given in listNums
    
    >>>square_sum([1, 2, 3])
    14
    """
    sum_ = 0
    for num in listNums:
        sum_ += num**2
    return sum_

ans = square_sum([1, 2, 3])
print(ans)
```

```python
nums = [1, 2, 3, 4, 5]
sq_sum = square_sum(nums)
print(sq_sum)
```
